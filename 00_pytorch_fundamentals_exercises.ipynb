### 3. Perform a matrix multiplication on the tensor from 2 with another random tensor with shape `(1, 7)` (hint: you may have to transpose the second tensor).

````python
# Create another random tensor
random_tensor_3 = torch.rand(1, 7)
print(random_tensor_3, random_tensor_3.shape)
# Perform matrix multiplication
# using @ for matrix multiplication
dot_product = random_tensor_3 @ random_tensor
print(f"\nUsing @  for matrix multiplication we have: \n{dot_product}")

# using tensor.mm for matrix multiplication
product = torch.mm(random_tensor_3, random_tensor)
print(f"\nUsing torch.mm for matrix multiplication we have: \n{product}")
````

### 4. Set the random seed to `0` and do 2 & 3 over again.


````python
# Set random seed


# Check for access to GPU
import torch
torch.cuda.is_available()

# Create two random tensors on GPU
torch.manual_seed(RANDOM_SEED_GPU)
tensor_A = torch.rand(2,3)
print(f"tensor_A = {tensor_A}, device = {device}")

torch.manual_seed(RANDOM_SEED_GPU)
tensor_B = torch.rand(2,3)
print(f"tensor_B = {tensor_B}, device = {device}")
````


### 7. Perform a matrix multiplication on the tensors you created in 6 (again, you may have to adjust the shapes of one of the tensors).

The output should look like:
```
(tensor([[0.3647, 0.4709],
         [0.5184, 0.5617]], device='cuda:0'), torch.Size([2, 2]))
```

````python
# Perform matmul on tensor_A and tensor_B
tensor_A_reshaped = tensor_A.reshape(3,2)
print(f"tensor_A_reshaped: \n{tensor_A_reshaped}")
Matmul_GPU = tensor_B @ tensor_A_reshaped
print(f"\n Matmul_GPU = \n{Matmul_GPU}, device = {device}, {Matmul_GPU.shape} ")
````

### 8. Find the maximum and minimum values of the output of 7.

````python
# Find max
max_value = torch.max(Matmul_GPU)
print(f"maximum value is: {max_value}")

# Find min
min_value = torch.min(Matmul_GPU)
print(f"minimum value is: {min_value}")
````

### 9. Find the maximum and minimum index values of the output of 7.

````python
# Find arg max
max_index = torch.argmax(Matmul_GPU)
print(f"maximum index value is: {max_index}")


# Find arg min
min_index = torch.argmin(Matmul_GPU)
print(f"minimum index value is: {min_index}")
````


### 10. Make a random tensor with shape `(1, 1, 1, 10)` and then create a new tensor with all the `1` dimensions removed to be left with a tensor of shape `(10)`. Set the seed to `7` when you create it and print out the first tensor and it's shape as well as the second tensor and it's shape.

The output should look like:

```
tensor([[[[0.5349, 0.1988, 0.6592, 0.6569, 0.2328, 0.4251, 0.2071, 0.6297,
           0.3653, 0.8513]]]]) torch.Size([1, 1, 1, 10])
tensor([0.5349, 0.1988, 0.6592, 0.6569, 0.2328, 0.4251, 0.2071, 0.6297, 0.3653,
        0.8513]) torch.Size([10])
```

````python
# Set seed
RANDOM_SEED = 42


# Create random tensor
torch.manual_seed(RANDOM_SEED)
tensor_C = torch.rand(1, 1, 1, 10)

# Remove single dimensions
tensor_C_squeezed = tensor_C.squeeze()

# Print out tensors and their shapes
print(f"The shape of tensor_C is: {tensor_C.shape}")
print(f"\nThe shape of tensor_C_squeezed is: {tensor_C_squeezed.shape}")
````
